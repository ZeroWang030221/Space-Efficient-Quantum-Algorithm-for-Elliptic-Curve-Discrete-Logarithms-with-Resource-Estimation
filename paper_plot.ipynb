```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

data = {
    512: 312086152,
    384: 176651379,
    256: 98484227,
    224: 75723823,
    192: 55948465,
    160: 39160053,
    128: 22137659,
    64: 5012016,
}

data = {k: v * 2 for k, v in data.items()}

n_vals = sorted(data.keys())
y_vals = [data[n] for n in n_vals]

plt.figure(figsize=(8, 5))
plt.plot(n_vals, y_vals, marker='o', linewidth=2)

plt.title("Toffoli-count vs n", fontsize=14)
plt.xlabel("n", fontsize=12)
plt.ylabel("Toffoli", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

for x, y in zip(n_vals, y_vals):
    plt.text(x, y * 1.02, f"n={x}", fontsize=9, ha='center')

ax = plt.gca()
y_min = min(y_vals)
y_max = max(y_vals)
plt.ylim(y_min * 0, y_max * 1.05)

ax.yaxis.set_major_locator(ticker.MultipleLocator(1e8))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(2.5e7))

ax.ticklabel_format(style='sci', axis='y', scilimits=(0, 0))

plt.tight_layout()

plt.savefig("toffoli.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
